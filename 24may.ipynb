```python
df = pd.read_csv('/kaggle/input/praactice/practice.csv', encoding='unicode_escape')
```

```python
x = df['Name ']
y=df['Marks']
```

```python
plt.plot(x,y,'r-.')
```

```python
plt.plot(x,y,'#05A4B3')
plt.ylabel('Marks')
plt.xlabel('Name')
plt.title('ADIT CLASS',{'color':'red','size':15,'family':'serif'})
plt.grid(axis='both',color='green', linestyle='-.',fillstyle='full',mouseover=True)
```

```python

plt.figure(figsize=(16,10))
plt.bar(x,y,color='#A0d0C0')
plt.ylabel('Marks')
plt.xlabel('Name')
plt.grid(axis='both',color='green', linestyle='-.',fillstyle='full',mouseover=True)


```
