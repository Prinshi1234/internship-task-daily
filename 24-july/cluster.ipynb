```python
import matplotlib.pyplot as plt
plt.scatter(df['Age'], df['Income'])
```

```python
from sklearn.cluster import KMeans
model = KMeans(n_clusters=3)
```

```python
# model = KMeans()
```

```python
model
```

```python
y_pred = model.fit_predict(df[['Age','Income']])
```

```python
y_pred
```

```python
df['cluster'] = y_pred
```

```python
df
```

```python
model.cluster_centers_
```

```python
lmc= df[df.cluster==0]
mc= df[df.cluster==1]
uc= df[df.cluster==2]
# df3= df[df.cluster==3]
# df4= df[df.cluster==4]
# df5= df[df.cluster==5]
# df6= df[df.cluster==6]
# df7= df[df.cluster==7]
```

```python
# Plot scatter plots
plt.scatter(lmc.Age, lmc.Income, label='LMC')
plt.scatter(mc.Age, mc.Income, label='MC')
plt.scatter(uc.Age, uc.Income, label='UC')
# plt.scatter(df3.Age, df3.Income)
# plt.scatter(df4.Age, df4.Income)
# plt.scatter(df5.Age, df5.Income)
# plt.scatter(df6.Age, df6.Income)
# plt.scatter(df7.Age, df7.Income)

# Plot cluster centers
plt.scatter(model.cluster_centers_[:,0], model.cluster_centers_[:,1], color='r', marker='*', label='Cluster Centers')

# Adding labels and title
plt.xlabel('Age')
plt.ylabel('Income')
plt.title('Age vs Income Scatter Plot with Cluster Centers')

# Adding a legend
plt.legend()

# Show the plot
plt.show()
```

```python
sse = []
sse_r = range(1, 10)

for i in sse_r:
    km = KMeans(n_clusters=i, random_state=42)  # It's a good practice to set a random_state for reproducibility
    km.fit(df[['Age', 'Income']])  # Use double brackets to select multiple columns
    sse.append(km.inertia_) 
```

```python
plt.plot(sse_r, sse, marker='o')
plt.xlabel('Number of clusters')
plt.ylabel('SSE')
plt.title('Elbow Method for Optimal Number of Clusters')
plt.show()
```
